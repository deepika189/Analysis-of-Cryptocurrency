```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import Normalizer
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import plotly.express as px #for plotting the scatter plot
```

```python
df= pd.read_excel('/content/Correlation_cryptos.xlsx')
```

```python
df
```

```python
dfNew=df.drop(['Date'], axis=1)
```

```python
dfNew
```

```python
fig = px.scatter_matrix(dfNew)
fig.update_layout(
    autosize=False,
    width=800,
    height=800,)
fig.show()
```

```python
corr = dfNew.corr()
plt.figure(figsize = (15,8)) #To set the figure size
sns.heatmap(data=corr, square=True , annot=True, cbar=True) 
```

```python
for column in dfNew.columns:
  dfNew[column]=dfNew[column]/dfNew[column].max()
```

```python
dfNew
```

```python
fig = px.scatter_matrix(dfNew)
fig.update_layout(
    autosize=False,
    width=800,
    height=800,)
fig.show()
```

```python
corr = dfNew.corr()
plt.figure(figsize = (15,8)) #To set the figure size
sns.heatmap(data=corr, square=True , annot=True, cbar=True) 
```

```python
spearman_corr = dfNew.corr(method="spearman")
plt.figure(figsize = (15,8)) #To set the figure size
sns.heatmap(data=spearman_corr, square=True , annot=True, cbar=True)
```

```python
Pearson_corr = dfNew.corr(method="pearson")
plt.figure(figsize = (15,8)) #To set the figure size
sns.heatmap(data=Pearson_corr, square=True , annot=True, cbar=True)
```

```python
Kendall_corr = dfNew.corr(method="kendall")
plt.figure(figsize = (15,8)) #To set the figure size
sns.heatmap(data=Kendall_corr, square=True , annot=True, cbar=True)
```
